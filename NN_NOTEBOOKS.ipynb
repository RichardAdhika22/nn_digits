* 神经网络 => 参考了人脑神经元的小部分机制
  * 人脑由很多神经元组成，神经元之间互相连接
  * 当一个神经元处于激活状态时，它会向相邻的神经元发送信号
* 神经网络由多个神经元组成
* 神经元的工作原理:
  * 每个神经元会接受到若干数据（输入数据）（输入数据数据可能来自其他神经元或样本数据集）
  * 每个神经元为每个数据都对应一个权重w。如果输入数据共512个，那么这个神经元就有512个w
  * 每个神经元的第一个工作：把每个输入数据与对应的w进行乘法，然后求总和，最后加上w0
  * 这个工作和回归算法很相似：G = X dot W + w0
  * 每个神经元的第二个工作：把第一步工作的结果，输入激活函数进行计算，其结果就是神经元的输出
  * 激活函数有很多种：最常用就是sigmoid函数：Z = 1 / (1 + e **(-G))
  * 神经元的输出会作为下一个神经元的输入
* 
* Neural Network => Refers to a small part of the mechanism of neurons in the human brain
   * The human brain is composed of many neurons, and the neurons are connected to each other
   * When a neuron is active, it sends a signal to neighboring neurons
* A neural network consists of multiple neurons
* How neurons work:
   * Each neuron will receive some data (input data) (input data may come from other neurons or sample data sets)
   * Each neuron corresponds to a weight w for each data. If there are a total of 512 input data, then this neuron has 512 w
   * The first job of each neuron: multiply each input data with the corresponding w, then sum, and finally add w0
   * This works very similar to the regression algorithm: G = X dot W + w0
   * The second job of each neuron: input the result of the first step into the activation function for calculation, and the result is the output of the neuron
   * There are many activation functions: the most commonly used is the sigmoid function: Z = 1 / (1 + e **(-G))
   * The output of a neuron will be used as the input of the next neuron

* 神经元总结
  * 输入n个样本，每个样本m个特征，那么单个神经元具有m个w和1个w0
  * 每个样本都会产生一个输出，n个样本 => 输出为(n, )
* 把若干个神经元按照层的形态进行组合，就是神经网络，也称为全连接神经网络
* 神经网络由若干层组成，每一层可以包含若干个神经元，神经元的数量用h1,h2.....表示
* 第一层网络
  * 一层神经网络中的每个神经元都会接收到相同的输入数据
  * 输入n个样本，每个样本m个特征，那么第一层神经网络中的每个神经元都具有m个w和一个w0，一共(h1, m)个w, h1个w0
  * G1 = X dot W.T + w0,  Z1 = sigmoid(G1)，Z1就是第一层网络的输出
  * G1和Z1的形状为(n, h1), 意思是n个样本，每个样本都通过h1个神经元输出了一个数值
  * 第一层网络的输出作为第二层网络的输入
* 第二层网络
  * 第一层n个样本每个样本都输出了h1个数值，这些数值作为第二层网络的输入
  * 第二层网络的每个神经元都有h1个w和1个w0，一共有(h2，h1)个w和h2个w0
  * G2 = Z1 dot W.T + w0, Z2 = sigmoid(G2),Z2是第二层网络的输出
  * G2和Z2的形状为(n, h2), 意思是n个样本，每个样本都通过h2个神经元输出了一个数值
  * 第二层网络的输出作为第三层网络的输入
* 整体网络的层数和每一层神经元的数量是由人来决定的，人决定的数据称为超参数
* 可以把神经网络看作是很多个回归算法+sigmiod组成，理论上它可以拟合任意复杂的函数
* 
* neuron summary
   * Input n samples, each sample has m features, then a single neuron has m w and 1 w0
   * Each sample produces an output, n samples => output is (n, )
* Combining several neurons according to the shape of the layer is a neural network, also known as a fully connected neural network
* The neural network consists of several layers, each layer can contain several neurons, and the number of neurons is represented by h1, h2.....
* First layer network
   * Each neuron in a layer of neural network will receive the same input data
   * Input n samples, each sample has m features, then each neuron in the first layer of neural network has m w and one w0, a total of (h1, m) w, h1 w0
   * G1 = X dot W.T + w0, Z1 = sigmoid(G1), Z1 is the output of the first layer network
   * The shape of G1 and Z1 is (n, h1), which means n samples, and each sample outputs a value through h1 neurons
   * The output of the first layer network is used as the input of the second layer network
* Layer 2 network
   * Each sample of n samples in the first layer outputs h1 values, which are used as the input of the second layer network
   * Each neuron in the second layer network has h1 w and 1 w0, and there are (h2, h1) w and h2 w0
   * G2 = Z1 dot W.T + w0, Z2 = sigmoid(G2), Z2 is the output of the second layer network
   * The shape of G2 and Z2 is (n, h2), which means n samples, and each sample outputs a value through h2 neurons
   * The output of the second layer network is used as the input of the third layer network
* The number of layers of the overall network and the number of neurons in each layer are determined by humans, and the data determined by humans is called hyperparameters
* The neural network can be regarded as a composition of many regression algorithms + sigmiod, theoretically it can fit any complex function

* 神经网络总结
  * 神经网络用神经元组成层，最终的目的是计算yp, 然后计算loss，并用梯度下降使loss最小
  * 输入n个样本，每个样本m个特征，这些样本经过若干层后，输出结果的形状为(n, hx),其中hx为最后一层网络的神经元数量
  * 这个输出不是yp，也不能用于计算loss
  * 这是神经网络的前半部分，它负责“学习”样本数据中的“规律”或特征
* 神经网络的后半部分按照最终完成的任务进行设计
* 最简单的任务：回归任务、分类任务
* 
* Neural Network Summary
   * The neural network uses neurons to form a layer. The ultimate goal is to calculate yp, then calculate loss, and use gradient descent to minimize loss
   * Input n samples, each sample has m features. After these samples pass through several layers, the output shape is (n, hx), where hx is the number of neurons in the last layer of the network
   * This output is not yp, nor can it be used to calculate loss
   * This is the first half of the neural network, which is responsible for "learning" the "regularities" or features in the sample data
* The second half of the neural network is designed according to the final completed task
* The simplest tasks: regression tasks, classification tasks

* 回归任务：每个样本输出一个数值，例如：房价
* 神经网络最终的输出是(n, 1)
* 解决方案：在前半部分的尾部再加一层神经网络，它的神经元数量只有1
* 神经网络的输出和Y计算MES LOSS
* 神经网络的输出就是n个样本的yp
* 
* Regression task: output a value for each sample, for example: house price
* The final output of the neural network is (n, 1)
* Solution: Add a layer of neural network at the end of the first half, and its number of neurons is only 1
* Neural network output and Y calculation MES LOSS
* The output of the neural network is the yp of n samples

* 分类任务：每个样本输出k个类别的概率
* 神经网络最终的输出是(n,k)
* 解决方案：在前半部分的尾部再加一层神经网络，它的神经元数量为k
* 神经网络的输出和one-hot矩阵计算Cross Entropy LOSS
* 神经网络的输出就是概率，yp = argmax(神经网络的输出)
* 
* Classification task: the probability of outputting k categories for each sample
* The final output of the neural network is (n,k)
* Solution: Add a layer of neural network at the end of the first half, and its number of neurons is k
* Neural network output and one-hot matrix calculation Cross Entropy LOSS
* The output of the neural network is the probability, yp = argmax (the output of the neural network)

```python
data = pd.read_csv("digit.csv")

print(data.shape)
```

```python
import torch.nn as nn

class MyNet(nn.Module):
    def __init__(self):
        noNeurons = 500
        
        super(MyNet, self).__init__()
        self.fc1 = nn.Linear(784, noNeurons) #..., no of neurons
        self.fc2 = nn.Linear(noNeurons,10)  #no of neurons, output
    
    def forward(self, x):
        x = x.view(-1,784) #-1 allows computer automatically determines the size
        x = nn.functional.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = MyNet().to(device=my_device)
```

* Pytorch如何读取数据
* Pytorch读取数据框架

* How Pytorch reads data
* Pytorch read data frame

```python
from torch.utils.data import DataLoader

class my_dataset(torch.utils.data.Dataset):
    def __init__(self, filepath):
        self.data = []
        dataPane1 = pd.read_csv(filepath)
        
        for i in range (len(dataPane1)):
            self.data.append((dataPane1.iloc[i,1:], dataPane1.iloc[i,0])) #(value, label), value is also called as feature


    def __getitem__(self, index):
        feature, label = self.data[index]
        return torch.tensor(feature, dtype= torch.float32), torch.tensor(label, dtype = torch.int64)

    def __len__(self):
        return len(self.data)

ds = my_dataset('digit.csv')
loader = DataLoader(ds, batch_size = 256, shuffle = True, drop_last=True)




```

* Pytorch高层API中的随机梯度下降算法 Stochastic Gradient Descent Algorithm in Pytorch High Level API
* optimizer = opt.SGD(model.parameters(), lr=learning_rate)
* optimizer.step()
* optimizer.zero_grad()

```python
import torch.optim as opt

optimizer = opt.SGD(model.parameters(), lr= 0.005)
for epoch in range (10):
   
    running_loss = 0.0

    for j, (X,Y) in enumerate(loader, 0):
        X = X.to(my_device)
        Y = Y.to(my_device)

        optimizer.zero_grad()
        
        outputs = model(X)
        loss = criterion(outputs, Y)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        if j % 100 == 99:
            print('[%d, %5d] loss: %.3f' % (epoch + 1, j +1, running_loss/100))
            running_loss = 0.0
        
        
print('Training Finished')

```
